```python
import nltk
nltk.download("punkt_tab")
nltk.download("averaged_perceptron_tagger_eng") #this file is imporatant to download for pos_tag
```

```python
from nltk.tokenize import word_tokenize
paragraph="""A paragraph is a distinct section of writing, usually starting on a new line with an indent,
that groups related sentences around a single central idea or topic, serving as a fundamental building
block for organizing longer texts like essays and articles for better readability and structure.
Typically, a paragraph includes a topic sentence, supporting sentences with details or examples, and
sometimes a concluding sentence that bridges to the next paragraph, helping readers follow the flow of thought. """

tokens=word_tokenize(paragraph)
print(tokens)
```

```python
pos_tags=nltk.pos_tag(tokens)
for word,tag in pos_tags:
    print(word," --> ",tag)
```
