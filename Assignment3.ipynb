```python
# create a corpus or sentence and convet it to lower case and then do tokenization.....
from nltk import ngrams
from nltk.tokenize import word_tokenize
corpus= """NLTK is a leading platform for building Python programs to work with human language data.
           It provides easy-to-use i
           nterfaces to over 50 corpora and lexical resources such as WordNet,
          along with a suite of text processing libraries for classification, tokenization, stemming, tagging,
          parsing, and semantic reasoning, wrappers for industrial-strength NLP libraries, and an active discussion forum."""
corpus= corpus.lower()

tokens=word_tokenize(corpus)
print(tokens)
```

```python
# Removing rhe Stopwords
from nltk.corpus import stopwords
nltk.download('stopwords')
stopword=list(stopwords.words('english'))
```

```python
# Lemmatization (Convert work in their true form)
nltk.download("wordnet")
from nltk.stem import WordNetLemmatizer
lemma=WordNetLemmatizer()
lemmatized_tokens=[lemma.lemmatize(token) for token in tokens if token not in stopword ]

print(lemmatized_tokens)
```

```python
bigrams= list(ngrams(lemmatized_tokens,2))
print(bigrams)
```

```python
trigrams=list(ngrams(lemmatized_tokens, 3))
print(trigrams)
```

```python
quadgrams=list(ngrams(lemmatized_tokens, 4))
print(quadgrams)
```

```python
def get_ngrams(n, tokens):
    ngrams = []
    tokens = (n-1)*['<START>']+tokens
    for i in range(n - 1, len(tokens)):
        context = tuple(tokens[i - p - 1] for p in reversed(range(n - 1)))
        target_word = tokens[i]
        ngrams.append((context, target_word))
    return ngrams
```
